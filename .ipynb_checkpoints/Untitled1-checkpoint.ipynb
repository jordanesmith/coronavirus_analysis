# Neatening the numpy arrays of data

Data is from table with regular titles at intervals, such as 'Date' repetition (for viewing purposes on wikipedia)

Had to remove this for ease of analysis.

```python
numpy_tables_withoutChina = pd.DataFrame.to_numpy(df_withoutChina,dtype='object')
numpy_tables_china = pd.DataFrame.to_numpy(df_china,dtype='object')

def remove_annoying_lines(array):
    counter = 0
    ignore_rows=[0,1,2]
    while counter < np.shape(array)[0]:
        row = array[counter]
        is_row_string = False
        if counter in ignore_rows:
            pass
        else:
#             for value in row:
#                 if type(value) == str:
#                     is_row_string = True
            if type(row[0]) == datetime:
                pass
            elif str(row[0])[0] == '2':
                    pass
            elif type(row[0]) == pd.Timestamp:
                pass
            else:
                is_row_string = True
            
            if is_row_string == True:
                array = np.delete(array, (counter), axis=0)
        counter +=1 
    
    return(array)

numpy_tables_withoutChina = remove_annoying_lines(numpy_tables_withoutChina)
numpy_tables_china = remove_annoying_lines(numpy_tables_china)
```

```python
print(numpy_tables_withoutChina)
print(numpy_tables_china)
```

```python
data = numpy_tables_withoutChina

index_countries = []
counter = 0 
include_china = False

for country in data[0]:
    if country in top_5_countries:
        index_countries.append((country,counter))
    elif country == 'United States':
        if 'USA' in top_5_countries:
            index_countries.append((country,counter))
    elif country == 'United':
        if 'UK' in top_5_countries:
            index_countries.append((country,counter))
    elif 'China' in top_5_countries:
        if include_china == False:
            print('---remember to include total chinese cases from different excel sheet---')
        include_china = True
        
    counter +=1
    


###delete this next two lines, just made to select countrys
index_countries = [8,44,57,72]
```

#  Data for China

China has a seperate wikipedia page therefore have to calculate values seperately

```python
yvals_china = []
dates_china = []
previous_date_point = 0
number_cases = 0
date=0

if include_china == True:
    for date_point in numpy_tables_china[3:]:
        date = date_point[0]
        if type(date) != str and type(date_point[32]) != float:
            number_cases = date_point[32] + previous_date_point
            previous_date_point = number_cases
            
            dates_china.append(date)
            yvals_china.append(number_cases)
            
###need to convert dates_china from timestamp type to datetime tipe for analysis later
for i,date in enumerate(dates_china):
    if type(date) == pd.Timestamp:
        dates_china[i] = date.to_pydatetime()
    else:
        print(type(date))
```

# Data for Rest of World


```python
previous_cases = np.zeros((len(numpy_tables_withoutChina[0])),dtype=int)
dates_ = np.zeros(shape=(1,len(numpy_tables_withoutChina[1])),dtype = 'object')
yvals_ = np.zeros(shape=(max(index_countries)+1,len(numpy_tables_withoutChina[1])),dtype = 'object')

for j,date_point in enumerate(numpy_tables_withoutChina):
    if type(date_point[0]) != float:
        if date_point[0][0] == '2':
            for letter in range(len(date_point[0])):
                if date_point[0][letter] == '.': 
                    date_point[0] = date_point[0][:letter]+'-'+date_point[0][letter+1:]
            for letter in range(1,len(date_point[0])-1):
                if date_point[0][letter-1] == '-' and date_point[0][letter+1] == '-':
                    date_point[0] = date_point[0][:letter] + '0' + date_point[0][letter:]
                    
            #need to solve issue where data with source link on wikipedia tables is ignored by this code
              
            date = datetime.strptime(date_point[0], '%y-%m-%d')
            number_cases = np.zeros((len(numpy_tables_withoutChina[0])))
            for i in index_countries:
                if type(date_point[i]) == int:
                    if type(previous_cases[i]) == np.int32:
                        number_cases[i] = int(date_point[i] + previous_cases[i])
                        previous_cases[i] = int(number_cases[i])

                    elif previous_cases[i] == 0:
                        number_cases[i] = date_point[i]
                        previous_cases[i] = int(number_cases[i])
                else:
                    number_cases[i] = previous_cases[i]
                yvals_[i][j] = number_cases[i]                                            
            dates_[0][j] = date 

dates_ = dates_[0]

rows_to_be_deleted = []
for i in range(len(numpy_tables_withoutChina)):
    for i,day in enumerate(dates_):
        if day == 0:
            rows_to_be_deleted.append(i)

dates_ = np.delete(dates_,rows_to_be_deleted,0)
yvals_ = np.delete(yvals_,rows_to_be_deleted,1)
print('date range :', dates_[0],'to: ',dates_[-1])

```
